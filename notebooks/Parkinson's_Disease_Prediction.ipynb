# Data Pre-Processing

```python
X = parkinsons_data.drop(columns = ['name','status'],axis=1)
y = parkinsons_data['status']
```

```python
X.head()
```

```python
y.head()
```

# Splitting the data to training data and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2,stratify=y)
```

```python
X.shape, X_train.shape, X_test.shape
```

# Data Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Model Training

## **Support Vector Machine**

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report
```

## For Training Data

```python
# Accuracy Score
X_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_pred = X_train_pred, y_true = y_train)
train_accuracy
```

```python
# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y_pred = X_train_pred, y_true = y_train)
```

```python
# Classification report
print(classification_report(y_pred = X_train_pred, y_true = y_train))
```

## For Test Data

```python
# Accuracy Score
X_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_pred = X_test_pred, y_true = y_test)
test_accuracy
```

```python
# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y_pred = X_test_pred, y_true = y_test)
```

```python
# Classification Report
print(classification_report(y_pred = X_test_pred, y_true = y_test))
```
